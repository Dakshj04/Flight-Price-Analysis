```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
df=pd.read_excel("flight_price.xlsx")
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
#feature engineering
df["Date"]=df["Date_of_Journey"].str.split("/").str[0]
```

```python
df["Month"]=df["Date_of_Journey"].str.split("/").str[1]
```

```python
df["Year"]=df["Date_of_Journey"].str.split("/").str[2]
```

```python
df.head()
```

```python
df["Date"]=df["Date"].astype(np.int64)
```

```python
df["Month"]=df["Month"].astype(np.int64)
```

```python
df["Year"]=df["Year"].astype(np.int64)
```

```python
df.info()
```

```python
df.drop("Date_of_Journey",axis=1,inplace=True)
```

```python
df.head()
```

```python
df["Arrival_Hour"]=df['Arrival_Time'].str.split(":").str[0]
```

```python
df.head()
```

```python
df['Arrival_Time']=df['Arrival_Time'].apply(lambda x:x.split(" ")[0])
```

```python
df["Arrival_Minutes"]=df['Arrival_Time'].str.split(":").str[1]
```

```python
df.head()
```

```python
df["Arrival_Hour"]=df["Arrival_Hour"].astype(np.int64)
df["Arrival_Minutes"]=df["Arrival_Minutes"].astype(np.int64)
```

```python
df.head()
```

```python
df.drop("Arrival_Time",axis=1,inplace=True)
```

```python
df.info()
```

```python
df.head()
```

```python
df["Dep_Hour"]=df['Dep_Time'].str.split(":").str[0]
df["Dep_Minute"]=df['Dep_Time'].str.split(":").str[1]
```

```python
df.drop("Dep_Time",axis=1,inplace=True)
```

```python
df.head()
```

```python
df["Duration"]=df["Duration"].apply(lambda x:x.replace('h','').replace('m',''))
```

```python
df["Duration_Hour"]=df['Duration'].str.split(" ").str[0]
df["Duration_Minute"]=df['Duration'].str.split(" ").str[1]
df.drop("Duration",axis=1,inplace=True)
```

```python
df.head()
```

```python

df["Total_Stops"].unique()
```

```python
df["Total_Stops"].mode()
```

```python
df[df["Total_Stops"].isnull()]
```

```python
df["Total_Stops"]=df["Total_Stops"].map({"non-stop":0,"1 stop":1,"2 stops":2,"3 stops":3,"4 stops":4,np.nan:1})
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop("Route",axis=1,inplace=True)
```

```python
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder=OneHotEncoder()
```

```python
encoder.fit_transform(df[["Airline","Source","Destination"]]).toarray()
```

```python
df1= pd.DataFrame(encoder.fit_transform(df[["Airline","Source","Destination"]]).toarray(),columns=encoder.get_feature_names_out())
```

```python
df=pd.concat([df,df1],axis=1)
```

```python
df.head()
```

```python
df.drop("Airline",axis=1,inplace=True)
```

```python
df.drop("Source",axis=1,inplace=True)
df.drop("Destination",axis=1,inplace=True)
```

```python
df.head()
```

```python
df.to_csv("cleaned_flight_data.csv", index=False)
```

```python


# Set seaborn style
sns.set_style("darkgrid")

# 1. Price Distribution
plt.figure(figsize=(8, 5))
sns.histplot(df["Price"], bins=30, kde=True, color='skyblue')
plt.title("Distribution of Flight Prices")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()


# 2. Total Stops vs Price
plt.figure(figsize=(8, 5))
sns.boxplot(x="Total_Stops", y="Price", data=df, palette="coolwarm")
plt.title("Effect of Total Stops on Price")
plt.xlabel("Total Stops")
plt.ylabel("Price")
plt.show()

# 3. Month-wise Average Price
plt.figure(figsize=(8, 5))
sns.barplot(x="Month", y="Price", data=df, palette="magma", ci=None)
plt.title("Month-wise Average Flight Price")
plt.xlabel("Month")
plt.ylabel("Average Price")
plt.show()

# 4. Departure vs Arrival Time Distribution
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

sns.histplot(df["Dep_Hour"], bins=24, kde=True, ax=ax[0], color='blue')
ax[0].set_title("Departure Hour Distribution")
ax[0].set_xlabel("Hour")

sns.histplot(df["Arrival_Hour"], bins=24, kde=True, ax=ax[1], color='red')
ax[1].set_title("Arrival Hour Distribution")
ax[1].set_xlabel("Hour")
plt.show()


# 5. Source vs. Price
plt.figure(figsize=(8, 5))
source_cols = ["Source_Chennai", "Source_Delhi", "Source_Kolkata", "Source_Mumbai"]
df_melted = df.melt(id_vars=["Price"], value_vars=source_cols, var_name="Source", value_name="Presence")
df_melted = df_melted[df_melted["Presence"] == 1]

sns.boxplot(x="Source", y="Price", data=df_melted, palette="Set2")
plt.title("Flight Source vs. Price")
plt.xlabel("Source City")
plt.ylabel("Price")
plt.xticks(rotation=45)
plt.show()

# 6. Destination vs. Price
plt.figure(figsize=(8, 5))
dest_cols = ["Destination_Banglore", "Destination_Cochin", "Destination_Delhi", 
             "Destination_Hyderabad", "Destination_Kolkata", "Destination_New Delhi"]
df_melted = df.melt(id_vars=["Price"], value_vars=dest_cols, var_name="Destination", value_name="Presence")
df_melted = df_melted[df_melted["Presence"] == 1]

sns.boxplot(x="Destination", y="Price", data=df_melted, palette="Set1")
plt.title("Flight Destination vs. Price")
plt.xlabel("Destination City")
plt.ylabel("Price")
plt.xticks(rotation=45)
plt.show()
```
